```python
!pip install matplotlib
!pip install SimpleITK
import matplotlib.pyplot as plt
import helpers as h
import numpy as np
from configs import *
```

```python
image_file_names = ["IMG-0002-00074.dcm", "IMG-0002-00110.dcm", "IMG-0012-00122.dcm"]
gt_file_names = ["liver_GT_074.png", "liver_GT_110.png", "liver_GT_122.png"]
configs = get_configs('all')
n_cluster = [4,8,12]
```

```python
for j in range(len(image_file_names)):
  print("image: ", image_file_names[j])
  image = h.get_image(image_file_names[j], configs[0])
  gt_image = h.get_ground_truth_image(gt_file_names[j], configs[0])

  for ii in range(len(n_cluster)):
    configs[0]['n_clusters']=n_cluster[ii]
    configs[1]['n_clusters']=n_cluster[ii]        
    configs[2]['n_clusters']=n_cluster[ii]
    configs[3]['n_clusters']=n_cluster[ii]

    similarity_matrices = []
    masks_withSpatial = []
    graphs_withSpatial = []
    avgs_cc_withSpatial= []
    nc_withSpatial= []
    infos = []
    dices = []
    lbls=[]
    hd_coefs = []

    for i in range(len(configs)):
        print('\n----------------------------------------------------------------')
        print(configs[i])
        print('----------------------------------------------------------------')
        similarity_matrices.append(h.build_weighMatrix(h.get_image(image_file_names[j], configs[i]), configs[i]))
        # sparsefy
        similarity_matrices[i] = h.sparsefy_simi_matrix(similarity_matrices[i], sparse_thresh=0.5)
        # build segmentation masks/cluster.labeles and graphs
        labels, _, _ = h.get_spectral_clustering(similarity_matrices[i], configs[i])
        masks_withSpatial.append(h.get_cluster_image(labels, image.shape))
        graphs_withSpatial.append(h.build_graph_of_simiMatrix(similarity_matrices[i]))
        # find best cluster
        best = (0,0,0)
        for l in labels:
          mask_for_label = h.get_mask_with_specLabel(mask=masks_withSpatial[i], label=l)
          dice_coeff = h.get_dice_coeff(mask_for_label, gt_image)
          hausdorff_dist = h.get_hausdorff_dist(mask_for_label, gt_image)
          if best[0] < dice_coeff:
            best = (dice_coeff, l, hausdorff_dist)
        
        dices.append(best[0])
        lbls.append(best[1])
        hd_coefs.append(best[2])

      # Get the average clustering coef and the normalized cut values for the cuts shown in the plot above

        avgs_cc_withSpatial.append(h.get_avg_cluster_coef(graphs_withSpatial[i]))

        nc_withSpatial.append(h.get_normalizedCut_value(graphs_withSpatial[i], 
                                                        h.get_subgrapg(graphs_withSpatial[i], np.where(masks_withSpatial[i]==best[1]))))
        infos.append(h.get_general_graph_info(graphs_withSpatial[i]))

    h.plot_multiple_masks(masks_withSpatial, lbls, image)
    print('Dice coefs: ', dices)
    print('hausdorffs distance: ', hd_coefs)
    print('\n informations: \n',infos)
    print('average clustering coefficient: ', avgs_cc_withSpatial)
    print('normalized cut for the graph: ', nc_withSpatial)
    print('lbls of clusters: \n', lbls)
```
